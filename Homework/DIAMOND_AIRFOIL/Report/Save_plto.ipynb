```python
# import necessary modules
from matplotlib import ticker
import sys
import os
import numpy as np
import subprocess as sbp
import time
from statistics import stdev
import matplotlib.pyplot as plt
from matplotlib import contour
```

```python
# Launch data


nh = 10
nH = 10

profile_h = np.logspace(start = 0, stop = -2, num = nh, base = 10)		# Mesh size at the profile buondary
freestream_H = np.logspace(start = 0.5, stop = -1, num = nH, base = 10)		# Mesh size at the freestream boundary
x, y = np.meshgrid(freestream_H, profile_h)
```

```python
# Import data 

Data = np.genfromtxt('file2', skip_header=True, delimiter = ',')

# Split and assign to variables

Time = Data[:,0]
Cd   = Data[:,1]
Cl   = Data[:,2]

# Reshape for plotting

Cd   = Cd.reshape(10,10)
Cd   = np.flip(Cd,0)
Cl   = Cl.reshape(10,10)
Cl   = np.flip(Cl,0)
```

```python
plt.contour(x,y,np.abs(Cd-0.11902025122033118), levels = 20, linewidths = 1)
plt.colorbar()
plt.xscale('log')
plt.yscale('log')
plt.title('CD computed - CD exact')
plt.xlabel('Freestream mesh dimension')
plt.ylabel('Profile mesh dimension')
plt.savefig('Cd.eps', transparent = True)
```

```python
plt.contour(x,y,np.abs(Cl-0.15583059627202586), levels = 20, linewidths = 1)
plt.colorbar()
plt.xscale('log')
plt.yscale('log')
plt.title('CL computed - CL exact')
plt.xlabel('Freestream mesh dimension')
plt.ylabel('Profile mesh dimension')
plt.savefig('Cl.eps', transparent = True)
```
